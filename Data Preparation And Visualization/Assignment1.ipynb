```python
import pandas
import numpy as np
import missingpy
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from missingpy import KNNImputer
from sklearn.preprocessing import Normalizer
```

```python
df=pandas.read_csv("share/A10_DataPreperationAndVisualization-master/DataPreperation/data/scrapped_data_large.csv")
```

```python
df
```

1. Preprocessing Data for slicing original data for players representing India at international level

```python
df=df.replace("-",np.nan)
df["first_class.highest"]=df["first_class.highest"].replace('\*','',regex=True).astype(float)
bool_series=df["teams"].str.startswith("India",na=False)
df1=df[bool_series]
```

a) Removing outliers of players playing for India at Local Level based on Domain knowledge

```python
df1.drop(df1[df1.teams=="Indian University Occasionals"].index,inplace=True)
df1.drop(df1[df1.teams=="India Revenue,"].index,inplace=True)
```

```python
df2=df1.iloc[:,4:]
df2.astype(float)
```

b)Filling N/A values with 0 and normalizing data as many zero values are present and fit a standard Gaussian Normal curve along the data.

```python
df2.fillna(0,inplace=True)
arr=df2.values
x_norm=arr[:,0:]
scaler=Normalizer().fit(x_norm)
normalizedx=scaler.transform(x_norm)
columns=['first_class.average','first_class.balls_faced','first_class.boundaries','first_class.catches_taken','first_class.fifties','first_class.highest','first_class.hundreds','first_class.inns','first_class.matches','first_class.not_outs','first_class.runs','first_class.sixes','first_class.strike_rate','first_class.stumps']
df2 = pandas.DataFrame(np.atleast_2d(normalizedx),columns=columns)
df2
corr=df2.corr()
```

Calculation of correlation of above data attributes(features) is checked and one of the two attributes are removed with correlation>=0.75 nearest 2 places

```python
corr
```

Dropping one of the higher correlated feature from the dataset

```python
df2.drop("first_class.boundaries",1)
df2.drop("first_class.highest",1)
```

2. Setting Up Dataset for getting top 3 features responsible for selection in Team India. Here two labels are given 0 and 1 specifying 1 for India, and 0 for rest of the teams: India U-19s,India A etc for classification and prediction.

```python
df3=df1.iloc[:,3:]
df3.fillna(0,inplace=True)
df3.drop("first_class.boundaries",1,inplace=True)
df3.drop("first_class.highest",1,inplace=True)
df3.loc[df3['teams'] == 'India,', 'teams'] = 1
df3.loc[df3['teams'] != 1, 'teams'] = 0
arr=df3.iloc[:,1:].values
x_norm=arr[:,0:]
scaler=Normalizer().fit(x_norm)
normalizedx=scaler.transform(x_norm)
columns=['first_class.average','first_class.balls_faced','first_class.catches_taken','first_class.fifties','first_class.hundreds','first_class.inns','first_class.matches','first_class.not_outs','first_class.runs','first_class.sixes','first_class.strike_rate','first_class.stumps']
df9 = pandas.DataFrame(np.atleast_2d(normalizedx),columns=columns)
df9
```

Using Recursive Feature Elimination for selecting Top 3 features using Logistic Regression.

```python
array=df9.values
x=array[:,0:].astype(float)
y=df3.iloc[:,0].values.astype(float)
model = LogisticRegression()
rfe = RFE(model,3)
fit = rfe.fit(x,y)
```

Getting Features based on Boolean Indexing: The top three features would be the ones which have ranking 1 i.e. first_class.balls_faced, first_class.runs, first_class.strike_rate.

```python
print("Num Features: %d"% fit.n_features_)
print("Feature Ranking: %s"% fit.ranking_)
print("Selected Features: %s"% fit.support_)
print(df3.columns)
```

3. Setting up DataFrame for KNN missing value imputation

```python
df4=df1.iloc[:,3:]
df4.drop("first_class.boundaries",1,inplace=True)
df4.drop("first_class.highest",1,inplace=True)
df5=df4.iloc[:,1:]
df5
```

Taking nearest neighbors in KNN as square root of datapoints to the nearest odd value.

```python
imputer = KNNImputer(n_neighbors=13, weights="uniform")
```

```python
arrayupd=imputer.fit_transform(df5.values)
arrayupd
```

```python
df4.loc[df4['teams'] == 'India,', 'teams'] = 1
df4.loc[df4['teams'] != 1, 'teams'] = 0
df4 = df4.reset_index(drop=True)
df4
```

```python
columns=['first_class.average','first_class.balls_faced','first_class.catches_taken','first_class.fifties','first_class.hundreds','first_class.inns','first_class.matches','first_class.not_outs','first_class.runs','first_class.sixes','first_class.strike_rate','first_class.stumps']
df6 = pandas.DataFrame(np.atleast_2d(arrayupd),columns=columns)
df6
```

```python
df7=pandas.concat([df4.iloc[:,0],df6],axis=1)
df7
```

Setting up the DataFrame values by normalizing the data with the imputed missing data which was done through KNN.
Using Logistic Regression in Recursive Feature Elimination as above.

```python
arr=df7.values
x_norm=arr[:,1:]
scaler=Normalizer().fit(x_norm)
normalizedx=scaler.transform(x_norm)
columns=['first_class.average','first_class.balls_faced','first_class.catches_taken','first_class.fifties','first_class.hundreds','first_class.inns','first_class.matches','first_class.not_outs','first_class.runs','first_class.sixes','first_class.strike_rate','first_class.stumps']
df7 = pandas.DataFrame(np.atleast_2d(normalizedx),columns=columns)
array=df7.values
x=array[:,0:].astype(float)
y=df4.iloc[:,0].values.astype(float)
model = LogisticRegression()
rfe = RFE(model,3)
fit = rfe.fit(x,y)
```

Getting Features based on Boolean Indexing: The top three features would be the ones which have ranking 1 i.e. first_class.balls_faced, first_class.matches, first_class.runs

```python
print("Num Features: %d"% fit.n_features_)
print("Feature Ranking: %s"% fit.ranking_)
print("Selected Features: %s"% fit.support_)
print(df7.columns)
```
