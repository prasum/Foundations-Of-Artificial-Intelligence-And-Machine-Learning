1 c) Compute Euclidean Distance between two observations

```python
import pandas
import numpy as np
liver_data_ignore_dose=liver_data.iloc[:,:-1]

def max_min_norm(a):
    liver_data_ignore_dose_result=a.copy()
    for i in range(1,3):
        for attribute_name in liver_data_ignore_dose.columns:
            max_value=liver_data_ignore_dose[attribute_name].max()
            min_value=liver_data_ignore_dose[attribute_name].min()
            liver_data_ignore_dose_result.loc[i,attribute_name]=(liver_data_ignore_dose.loc[i,attribute_name]-min_value)/(max_value-min_value)
    return liver_data_ignore_dose_result

liver_data_ignore_dose_normalised=max_min_norm(liver_data_ignore_dose)
print(liver_data_ignore_dose_normalised)

o1=pandas.to_numeric(liver_data_ignore_dose_normalised.iloc[0,:],errors='coerce')
o2=pandas.to_numeric(liver_data_ignore_dose_normalised.iloc[1,:],errors='coerce')

euclideanDistance=np.sqrt(np.sum((o1-o2)**2))
print(euclideanDistance)
```

2 a) Compute variance of each row of dataset

```python
mRNA_subset=pandas.read_csv('data_mRNA_subset.txt',sep='\t')
mRNA_subset_ignore_first_column=mRNA_subset.iloc[:,1:]
var_probe_id=mRNA_subset_ignore_first_column.var(axis=1)
if (var_probe_id.iloc[:] <= 0.2).any():
    mRNA_subset_updated=mRNA_subset_ignore_first_column.drop(var_probe_id[var_probe_id.iloc[:] <= 0.2].index)
    print(mRNA_subset_updated)
else:
    print("no such probe-id is found!!!")
```

b) compute Spearman coefficient between first two observations

```python
mRNA_subset_ignore_first_column_rank=mRNA_subset_ignore_first_column.rank()
print(mRNA_subset_ignore_first_column_rank)
r1=pandas.to_numeric(mRNA_subset_ignore_first_column_rank.iloc[0,:],errors='coerce')
r2=pandas.to_numeric(mRNA_subset_ignore_first_column_rank.iloc[1,:],errors='coerce')
d=r1-r2
sum_dsquare=np.sum(d**2)
column_count=len(mRNA_subset_ignore_first_column_rank.columns)
spearman_coeff_corr=1-(6*sum_dsquare)/(column_count*(column_count-1)*(column_count+1))
spearman_coeff_corr
```
