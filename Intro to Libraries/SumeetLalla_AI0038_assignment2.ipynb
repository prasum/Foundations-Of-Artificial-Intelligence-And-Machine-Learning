```python
# Question 1
# Create a tensor of shape (2,3)  having normal distribution having mean 4 and std 0.5
# Create a tensor Y of shape (2,3) having a sequence of numbers 12,15,18,...., 150
# Get and Print sum of these two Tensors that is Z=  X+Y   (Shape of Z would be same as that of X or Y)
# For tensor Y it has sequence of size 47 and nearest greatest shape is 48 so the unfilled element is assumed to be the last sequence element

import tensorflow as tf
import numpy as np
X=tf.random.truncated_normal(
    [2,3],
    mean=4.0,
    stddev=0.5,
    dtype=tf.dtypes.float32)
Y=tf.constant(np.arange(12,153,3),dtype=tf.dtypes.float32,shape=[8,2,3])
#sequence_tensor_reshaped=tf.reshape(sequence_tensor,[-1,2,3])
Z=tf.add(X,Y)
model = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(model)
    print(sess.run(X))
    print(sess.run(Y))
    print(sess.run(Z))
sess.close()
tf.reset_default_graph()
```

```python
# Question 2
#Get a new tensor W = tanh(Z)
#Get a new tensor T= sigmoid(Z)
import tensorflow as tf
import numpy as np
X=tf.random.truncated_normal(
    [2,3],
    mean=4.0,
    stddev=0.5,
    dtype=tf.dtypes.float32)
Y=tf.constant(np.arange(12,153,3),dtype=tf.dtypes.float32,shape=[8,2,3])
#sequence_tensor_reshaped=tf.reshape(sequence_tensor,[-1,2,3])
Z=tf.add(X,Y)
T=tf.nn.sigmoid(Z)
W=tf.nn.tanh(Z)
model = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(model)
    print(sess.run(W))
    print(sess.run(T))
sess.close()
tf.reset_default_graph()
```

```python
## Question 3
# Replacing sigmoid/tanh with another(new) activation function that is ( exp^(f) - exp^(-f) )/( exp^(f) + exp^(-f) ) + (1 / ( 1 + exp^(-f) ))
import tensorflow as tf
W = tf.random_normal(shape=(3,2))
X = tf.constant([2,4], dtype=tf.float32)
B = tf.zeros(shape=(2))
f = W*X + B
newActivationResult = ((tf.exp(f) - tf.exp(-f)) / (tf.exp(f) + tf.exp(-f))) + (1 / ( 1 + tf.exp(-f)))
with tf.Session() as sess:
    output = sess.run(newActivationResult)
    print(output)


#If we want to replace new activation function with sigmoid, which line is to be changed? 

# The line with newActiation Result has to be changed. The below code is edited in the next cell.
```

```python
# Question 5
# Build an image(digit) classifier using keras for MNIST data using adagrad, 'binary_crossentropy' using MLP (not CNN)

# The MLP model is defined below
```

```python
import numpy as np
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.utils import np_utils
(xtrain, ytrain), (xtest, ytest) = mnist.load_data()
num_pixels = xtrain.shape[1] * xtrain.shape[2]
xtrain = xtrain.reshape(xtrain.shape[0], num_pixels).astype('float32')
xtest = xtest.reshape(xtest.shape[0], num_pixels).astype('float32')
xtrain = xtrain / 255
xtest = xtest / 255
ytrain = np_utils.to_categorical(ytrain)
ytest = np_utils.to_categorical(ytest)
num_classes = ytest.shape[1]
model = Sequential()
model.add(Dense(num_pixels, input_dim=num_pixels, kernel_initializer='normal', activation='relu'))
model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))
model.compile(loss='binary_crossentropy', optimizer='adagrad', metrics=['accuracy'])
model.fit(xtrain, ytrain, validation_data=(xtest, ytest), epochs=10, batch_size=200, verbose=2)
scores = model.evaluate(xtest, ytest, verbose=0)
print("Error: %.2f%%" % (100-scores[1]*100))
```
